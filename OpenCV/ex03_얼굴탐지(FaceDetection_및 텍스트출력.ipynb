```python
import cv2
import cvlib as cv #   ()
import matplotlib.pyplot as plt

img = cv2.imread('data/GettyImages-537449915-03b4a08b21a04d18b3462a5e9a910471.jpg')
faces, conf = cv.detect_face(img)
# detect_face()   
# faces :    [x1( x), y1( y), x2( x), y2( y)]
# confidence(conf) :  dl ' ' (0~1 )

#     
#   : (B, G, R) = (0, 255, 0)
line_color = (0, 255, 0)
#   ( )
line_thick = 2
#   
font = cv2.FONT_HERSHEY_PLAIN # SIMPLEX, DUPLEX, COMPLEX 

#     
# faces, conf ->         zip()   
for face, c in zip(faces, conf):

    # faces [x1, y1, x2, y2]  
    # x1, y1 :   
    # x2, y2 :   
    x1, y1, x2, y2 = face

    #       .
    if c < 0.6:
        continue

    #       
    # cv2.rectagle(, (x, y), (x, y), , )
    cv2.retangle(img, (x1, y1), (x2, y2), line_color, line_thick)

    #    ()  
    cv2.putText(
        img, #    
        f'{c:.2f}', #  (:.2f      )
        (x1, y1 - 10), #   10   
        font, #  
        1.2, #  (scale)
        line_color, #  
        line_thick #  
    )

cv2.imshow('box_family', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
