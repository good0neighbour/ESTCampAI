```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# opencv 활용 이미지 띄우기 및 마우스 클릭 확인하기
img = cv2.imread('data/trollge.jpg')
cv2.imshow('img', img)

def mouse_click(event, x, y, flags, param):
    # event : 마우스로 어떤 행동을 했는지
    # (x, y) : 마우스로 행동한 순간의 좌표
    # flags : 마우스 버튼의 상태 정보(왼쪽 버튼을 누른 상태인지, Ctrl, Shift, Alt 키가 눌려있는지)
    # param : 사용자 지정 변수(원하는 값을 전달가능)
    if event == cv2.EVENT_LBUTTONDOWN:
        print(x, y, '왼쪽 마우스 클릭')

cv2.namedWindow('img')
cv2.setMouseCallback('img', mouse_click)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('data/newspaper.jpg')

# 출력될 최종 이미지 크기 width, height
width, height = 640, 240 # 가로, 세로

# 접근 좌표 정리하기 : 좌상, 우상, 좌하(시계방향 4개지점(x, y)) > 좌표값을 리스트 담기
xy_list = []
def mouse_click(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN:
        # 마우스클릭을 통해 받아온 x, y 좌표를 [x, y] 리스트로 패킹해서 append
        xy_list.append([x, y])
        # 4개가 선택되면 자동으로 창 닫힘
        if len(xy_list) == 4:
            cv2.destroyWindow('img') # 원래 창 닫기
            show_result() # 변환 실행

def show_result():
    # 사용자가 찍은 원본 이미지 4개 좌표
    src = np.array(xy_list, np.float32)
    # 변환될 새로운 평면의 4개 좌표(좌상 -> 우상 -> 우하 -> 좌하)
    dst = np.array([[0, 0], # 좌상
                    [width, 0], # 우상
                    [width, height], # 우하
                    [0, height] # 좌하
                   ], np.sloat32)
    # 투시변환 행렬 계산(4개의 점을 기반으로 변환 규칙 생성)
    matrix = cv2.getPerspectiveTransform(src, dst) # getPerspectiveTransform(사용자좌표, 새로운평면좌표)
    # 변환 규칙을 이용해 이미지 전체를 새 시점으로 변환
    result = cv2.warpPerspective(img, matrix, (width, height))
    # 결과 이미지 출력
    cv2.imshow('result_img', result)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

cv2.imshow('img', img)
cv2.namedWindow('img')
cv2.setMouseCallback('img', mouse_click)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
