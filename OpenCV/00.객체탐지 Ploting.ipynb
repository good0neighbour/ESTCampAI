```python
import cv2
from ultralytics import YOLO
```

```python
# 모델 로드
model = YOLO("./data/best.pt")
```

```python
# 에측에  사용할 영상의 경로 지정
video_path = "./data/(객체탐지 테스트용) 호랑나비 영상.mp4"
```

```python
# 영상 객체 초기화
cap = cv2.VideoCapture(video_path)

# 영상 파일이 열려있는 반복
while cap.isOpened() :

    # 한 프레임 씩 읽어들이기
    succes, frame= cap.read() # 성공여부와 이미지 소스 반환

    # 프레임 읽기에 성공한 경우
    if succes :
        
        # 모델 예측 시도
        results = model(frame, conf = 0.6)

        # 바운딩박스가 표시된 plot 추출
        annotated_frame = results[0].plot()

        # OpenCV 원도우 출력
        cv2.imshow("YOLO Inference", annotated_frame)

        # 'q'를 입력하면 종료
        if cv2.waitKey(1) == ord("q") :
            print("사용자에 의한 반복문 종료")
            break
    else : # 프레임 읽기에 실패한 경우 (프레임을 다 읽은 경우)
        print("더 이상 읽을 프레임이 없음")
        break

# 영상파일 닫고, 리소스 해제
cap.release()

# OpenCV 윈도우 닫기
cv2.destroyAllWindows()
```
