### 동영상 출력하기
- cv2.waitKey(ms(밀리초)) : 동영상이 실행되는 시간에 대해서 제어
- cv2.waitKey() : 사용자의 키보드로부터 입력된 아스키 코드값을 반환 -> 윈도우창 꺼지는 기능 제어

```python
import cv2

cap = cv2.VideoCapture('data/dogs.mp4')
# 동영상을 가져올 때 VideoCapture(경로)
# 카메라를 사용할 때 VideoCapture(카메라번호)

while True:
    ch, frame = cap.read() # ch : 성공여부, frame : 받아온 이미지(프레임)

    # 실패할 경우 while 멈추기
    if not ch:
        print('동영상 재생 끝 or 읽기 실패')
        break
    # 제대로 읽어진 경우
    cv2.imshow('video', frame)
    # q 키를 입력했을 때 동영상이 꺼지게 하기
    key = cv2.waitKey(33) # 동영상 실행시간 제어(33ms -> 30프레임)
    # 키를 눌렀을 때 반환되는 아스키코드 값을 key 변수에 담아주기
    if key == ord('q'): # ord('q') 대신에 숫자 113을 사용해도 됨
        print('사용자에 의해 영상이 종료')
        break
cap.release() # 카메라 연결 자원 해제
cv2.destroyAllWindows()
```

```python
# esc를 눌렀을 때 카메라가 종료되게 실행해보기 - ord 쓰지않고
while True:
    ch, frame = cap.read() # ch : 성공여부, frame : 받아온 이미지(프레임)

    # 실패할 경우 while 멈추기
    if not ch:
        print('동영상 재생 끝 or 읽기 실패')
        break
    # 제대로 읽어진 경우
    cv2.imshow('video', frame)
    # q 키를 입력했을 때 동영상이 꺼지게 하기
    key = cv2.waitKey(33) # 동영상 실행시간 제어(33ms -> 30프레임)
    # 키를 눌렀을 때 반환되는 아스키코드 값을 key 변수에 담아주기
    if key == 27: # ord('q') 대신에 숫자 113을 사용해도 됨
        print('사용자에 의해 영상이 종료')
        break
cap.release() # 카메라 연결 자원 해제
cv2.destroyAllWindows()
```

### 동영상에서 이미지 캡쳐하기

```python
cap = cv2.VideoCapture('data/dogs.mp4')
cnt = 0
while True:
    ch, frame = cap.read()

    if not ch:
        print('동영상 재생 끝 or 읽기 실패')
        break

    cv2.imshow('video', frame)
    key = cv2.waitKey(33)
    if key == ord('q'):
        print('사용자에 의해 영상 종료')
        break
    # 사진 찍기, 프레임 저장하기
    if key == ord('w'):
        cnt += 1
        cv2.imwrite(f'data/dog_img{cnt}.png', frame, params = [cv2.IMWRITE_PNG_COMPRESSION, 0])
        # 압축수준(0~9) : 0 - 압축안함, 1~8 평균, 9 최대(용량 가장 적음, 저장 느림)
        print(f'dog_img{cnt} 사진 저장 완료')
cap.release()
cv2.destroyAllWindows()
```

#### 휴대폰으로 주변영상을 촬영하여 이미지 캡쳐하고(4장) plt 사용하여 액자형태로 출력해보기

- 흑백모드 : gray변경 필요, 흑백인지 여부를 판별하기 위한 변수 필요

1. 흑백인지 여부를 판별하기 위한 변수 필요
2. 흑백 모드를 적용하기 위한 별도의 frame을 구성(변수를 활용한 판별)
3. 버튼을 눌렀을 때 변수의 상태가 생기게끔

```python
cap = cv2.VideoCapture('data/video1.mp4')
cnt = 0
is_gray = False # 흑백 모드 여부

while True :
    ch, frame = cap.read()

    if not ch :
        print('동영상 재생 끝 or 읽기 실패')
        break

    # 흑백 화면 적용
    if is_gray == True:
        frame = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)

    cv2.imshow('video',frame)
    key = cv2.waitKey(33)

    # 흑백 모드 변경
    if key == ord('b'):
        if is_gray:
            is_gray = False
        else :
            is_gray = True

    # 일시정지
    if key == ord(' '):
        # 영상 정지
        while True:
            key2 = cv2.waitKey(0)
            # 영상 다시 실행
            if key2 == ord(' '):
                break
            if key2 == ord('q'):
                cap.release()
                cv2.destroyAllWindows()
    
    if key == ord('q'):
        print('사용자에 의해 영상 종료')
        break
    # 사진 찍기, 프레임 저장하기
    if key == ord('w') :
        cnt += 1
        cv2.imwrite(f'data/man_img{cnt}.png',frame,params=[cv2.IMWRITE_PNG_COMPRESSION,0])
        # 압축수준(0~9) : 0 - 압축안함, 1~8 평균, 9 최대(용량 가장 적음, 저장 느림)
        print(f'man_img{cnt} 사진 저장 완료!')
cap.release()
cv2.destroyAllWindows()
```
