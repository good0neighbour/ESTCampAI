### 오늘의 학습 목표
- ZDnet 뉴스사이트의 최신뉴스 제목과 링크를 수집하여 DataFrame으로 만든 후 csv파일로 저장하자
- https://zdnet.co.kr/news/?lstcode=0000&page=1
- 순서
  1. url주소를 통해 page정보 받아오기(요청)
  2. 응답을 받은 Str형태의 웹 문서를 html형태로 파싱하기
  3. 뉴스제목, 링크 리스트로 저장
  4. DataFrame으로 생성한 후 csv파일로 저장하기

```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
url = "https://zdnet.co.kr/news/?lstcode=0000&page=1"
res = req.get(url)
soup = bs(res.text, "lxml")
```

```python
# 뉴스 제목 수집
title = soup.select('div.assetText > a > h3')
```

```python
# 뉴스 제목에 해당하는 링크 수집
aTag = soup.select('div.news_box > div.newsPost > div.assetText > a')
```

```python
#개수 겁증
print(len(title), len(aTag))
```

```python
title_list = []
for i in title:
    title_list.append(i.text)
title_list
```

```python
aTag_list = []
for i in aTag:
    aTag_list.append('https://zdnet.co.kr' + i.get('href')) # i['href']
aTag_list
```

```python
# 뉴스 제목과 링크를 하나의 리스트로 저장
news_list = []
for i in range(len(title)):
    news_list.append([title[i].text, aTag[i].get("href")])
news_list
```

### 10개 페이지의 뉴스제목과 링크를 수집(페이지를 넘겨서)

```python
# url에서 변경되는 부분에 집중하자 -> url안에서 변수를 삽입해야겠구나 -> 포멧팅을 써보자
news_list2 = []
for i in range(1, 11):
    url = f"https://zdnet.co.kr/news/?lstcode=0000&page={i}"
    res = req.get(url)
    soup = bs(res.text, "lxml")

    title = soup.select('div.assetText > a > h3')
    link = soup.select('div.news_box > div.newsPost > div.assetText > a')

    for i in range(len(title)):
        news_list2.append([title[i].text, 'https://zdnet.co.kr' + link[i].get('href')])
```

```python
len(news_list2)
```

```python
import pandas as pd
data = pd.DataFrame(news_list2, columns = ["뉴스제목", "링크"])
```

```python
data.to_csv("뉴스데이터.csv", encoding = "utf-8-sig")
```
