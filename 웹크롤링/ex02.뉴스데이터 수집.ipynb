```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
# 링크 가져와서 요청 후 서버에게 응답받기
res = req.get("https://zdnet.co.kr/news/?lstcode=0070&page=1")
```

```python
# html 형식으로 파싱
soup = bs(res.text, "lxml")
```

```python
# select
title = soup.select("div.assetText > a > h3")
```

```python
for i in title:
    print(i.text)
```

```python
aTag = soup.select("div.news_box > div.newsPost > div.assetText > a")
```

```python
aTag[0]["href"]
```

```python
url_list = []
for i in aTag:
    print(i["href"])
    url_list.append(i["href"])
```
