## Selenium Library
- 수집 뿐만 아니라, 브라우저의 동작을 제어할 수 있는 자동화 도구

> ex) 마우스 클릭, 키보드 입력값을 전달, 스크롤 하강, 창 전환 등

- 동적인 컨텐츠(페이지를 얾겨다니거나, action이 필요할 때 등)를 수집할 때 주로 활용

```python
# 설치
# !pip install selenium
```

```python
# 필요한 도구 불러오기
from selenium import webdriver as wb # 컴퓨터 전용 브라우저를 실행&관리하는 도구
from selenium.webdriver.common.by import By # 요소 지정 방식을 제공하는 도구
from selenium.webdriver.common.keys import Keys # 키보드 입력을 전달하기 위한 도구
```

```python
# 컴퓨터 전용 브라우저 열어주기
driver = wb.Chrome()
```

```python
# url 요청 - 네이버
driver.get("https://www.naver.com")
```

```python
# 검색란 찾기
search = driver.find_element(By.ID, "query") # 경로 지정 방식,

# bs에서 요소 하나 찾고자 할 때 → select_one() → 셀레니움에서는 find_element()
# bs에서 요소 여러개 찾고자 할 때 → select() → 셀레니움에서는 find_elements()
```

```python
# 검색어 입력 (키보드 입력값 전달)
search.send_keys("체인소맨 레제")
```

```python
# ENTER 전달
search.send_keys(Keys.ENTER)
```

```python
# 뒤로가기
driver.back()
```

```python
# 검색어 입력 (키보드 입력값 전달)
# search.send_keys("체인소맨 레제")

# Error 발생 → 페이지가 새롭게 렌더링(ex.뒤로가기, 페이지 이동 등) 될 때마다
# 항상 요소(ex.검색란)을 새롭게 지정해서 가져와야
```

```python
# 검색란 찾기부터 다시
search = driver.find_element(By.ID, "query")
```

```python
# 다시 검색어 입력
search.send_keys("체인소맨 레제")
```

```python
# 돋보기 버튼 찾기
btn = driver.find_element(By.CSS_SELECTOR, "#sform > fieldset > button")
```

```python
# 클릭
btn.click()
```

```python
# 자원반납 (브라우저 닫기)
driver.quit() # 전체 종료
# close() # 탭 닫기
```
