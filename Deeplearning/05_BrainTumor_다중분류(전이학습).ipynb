```python
# EfficientNetV2S Î™®Îç∏ÏùÑ Ï¥àÍ∏∞ÌôîÎ•º ÌïòÍ≥† Î≥ÄÏàò Ï†ÄÏû•
# Î™®Îç∏Ïù¥ ÏÇ¨Ïö©ÌïòÎäî Í∞ÄÏ§ëÏπò : ÏÇ¨Ï†Ñ ÌïôÏäµÎêú Í∞ÄÏ§ëÏπò
# Î™®Îç∏Ïùò Î∂ÑÎ•òÎ∂ÄÎäî ÏÇ¨Ïö©ÌïòÏßÄ ÏïäÏùå.
# ÌäπÏÑ± Ï∂îÏ∂úÎ∂ÄÎßå Î∂àÎü¨ÏÑú ÏÇ¨Ïö©
image_prepro = EfficientNetV2S(
    include_top = False, # Î™®Îç∏Ïùò Î∂ÑÎ•òÎ∂Ä(ÏµúÏÉÅÏ∏µ)Îäî Ï†úÏô∏ÌïòÍ≥† Î™®Îç∏ Î°úÎìú
    weights = "imagenet", # ÏÇ¨Ï†Ñ ÌïôÏäµÎêú Í∞ÄÏ§ëÏπò ÏÇ¨Ïö©
    input_shape = (224, 224, 3), # ÏûÖÎ†• Ïù¥ÎØ∏ÏßÄ ÌÅ¨Í∏∞ : (224, 224), RGB Ï±ÑÎÑê Ïàò : 3
    pooling = "max" # ÏµúÏ¢Ö Ï∂úÎ†•ÏóêÏÑú Global Max Pooling ÏÇ¨Ïö©Ìï¥ Ï∂úÎ†• ÏïïÏ∂ï
)

image_prepro.summary()
```

```python
from tensorflow.keras.utils import plot_model
plot_model(image_prepro)
# Î™®Îç∏Ïùò Í∞Å Ï∏µÏùÑ Ïó∞Í≤∞Ìïú ÎÑ§Ìä∏ÏõåÌÅ¨ Îã§Ïù¥Ïñ¥Í∑∏Îû®ÏùÑ Î≥¥Ïó¨Ï§å
```

```python
# 1. ÎºàÎåÄ Íµ¨ÏÑ±
tumor_model = Sequential()

# 2. ÌäπÏÑ± Ï∂îÏ∂úÎ∂Ä
tumor_model.add(image_prepro)

# 3. Î∂ÑÎ•òÎ∂Ä(MLP)
# 1) 1Ï∞®Ïõê Î≥ÄÌôòÏ∏µ
tumor_model.add(Flatten())

# 2) 2048, relu
tumor_model.add(Dense(units = 2048, activation = "relu"))
# 3) 1024, relu
tumor_model.add(Dense(units = 1024, activation = "relu"))
# 4) 512, relu
tumor_model.add(Dense(units = 512, activation = "relu"))
# 5) dropout 0.2 ‚Üí ÎûúÎç§ÌïòÍ≤å 20%Ïóê Ìï¥ÎãπÌïòÎäî Îâ¥Îü∞ ÎπÑÌôúÏÑ±Ìôî
tumor_model.add(Dropout(0.2))
# 4. Ï∂úÎ†•Ï∏µ -> Îã§Ï§ëÎ∂ÑÎ•ò
tumor_model.add(Dense(units = 4, activation = "softmax"))
```

```python
# 4. ÌïôÏäµ Î∞©Î≤ï ÏÑ§Ï†ï
tumor_model.compile(
    loss = "categorical_crossentropy", # Ïõê-Ìï´ Ïù∏ÏΩîÎî© Ï†ÅÏö© ÏôÑÎ£å, Ïõê-Ìï´ Ïù∏ÏΩîÎî©Ïù¥ Î∂àÍ∞ÄÎä•Ìïú Îãµ : sparse_categorical_crossentropy
    optimizer = "adam",
    metrics = ["accuracy"]
)
```

```python
# Í≥ºÎåÄÏ†ÅÌï© Ï†úÏñ¥Î•º ÏúÑÌï¥ Ï°∞Í∏∞ÌïôÏäµ Ï§ëÎã® ÌôúÏö©
early = EarlyStopping(monitor = "val_accuracy", patience = 5, verbose = 1)
# monitor : Í∏∞Ï§Ä ÏÑ§Ï†ï, val_accuracyÍ∞Ä Ïù¥Ï†Ñ ÏµúÍ≥†Í∞íÎ≥¥Îã§ ÎÇÆÏúºÎ©¥ Í∞úÏÑ† ÏóÜÏùåÏúºÎ°ú Î¥Ñ ‚Üí patience_count + 1
# verbose : ÌïôÏäµ ÌòÑÌô©ÏùÑ ÏûêÏÑ∏Ìûà Ï∂úÎ†•ÌïòÍ≤†Îã§.
# 6. ÌïôÏäµ
h = tumor_model.fit(X_train, y_train,
                    validation_split = 0.2,
                    epochs = 100,
                    callbacks = [early],
                    batch_size = 32
                    )
```
