```python
import requests
from dataclasses import dataclass
from datetime import datetime

@dataclass
class Weather:
    city: str
    temp_c: float
    wind_kmh: float
    is_rainy: bool
    code: int

def season_by_month(month: int) -> str:
    if month in [12,1,2]:
        return "겨울"
    if month in [3,4,5]:
        return "봄"
    if month in [6,7,8]:
        return "여름"
    return "가을"

def geocode_city(city: str):
    url = "https://geocoding-api.open-meteo.com/v1/search"
    params = {"name": city, "count": 1, "language": "ko", "format": "json"}
    r = requests.get(url, params=params, timeout=12)
    r.raise_for_status()
    data = r.json()
    if not data.get("results"):
        return None
    res = data["results"][0]
    return res["latitude"], res["longitude"], res.get("name", city)

def fetch_weather(city: str) -> Weather | None:
    geo = geocode_city(city)
    if not geo:
        return None
    lat, lon, city_name = geo
    url = "https://api.open-meteo.com/v1/forecast"
    params = {
        "latitude": lat,
        "longitude": lon,
        "current": "temperature_2m,weather_code,wind_speed_10m",
        "timezone": "Asia/Seoul",
    }
    r = requests.get(url, params=params, timeout=12)
    r.raise_for_status()
    data = r.json()
    cur = data.get("current", {})
    temp = float(cur.get("temperature_2m"))
    wind = float(cur.get("wind_speed_10m"))
    code = int(cur.get("weather_code"))
    rainy_codes = {51,53,55,56,57,61,63,65,66,67,71,73,75,77,80,81,82,85,86,95,96,99}
    return Weather(
        city=city_name,
        temp_c=temp,
        wind_kmh=wind,
        is_rainy=(code in rainy_codes),
        code=code
    )

def temp_band(temp_c: float) -> str:
    if temp_c <= 0: return "영하~0도"
    if temp_c <= 8: return "1~8도"
    if temp_c <= 16: return "9~16도"
    if temp_c <= 22: return "17~22도"
    if temp_c <= 27: return "23~27도"
    return "28도 이상"

def outfit_reco(temp_c: float, is_rainy: bool, style: str):
    band = temp_band(temp_c)
    base, outer, shoes, acc = [], [], [], []

    if band == "영하~0도":
        outer = ["롱패딩/다운", "헤비 코트"]
        base = ["두꺼운 니트", "기모 이너", "울 팬츠/기모 데님"]
        shoes = ["방한 부츠", "두꺼운 양말"]
        acc = ["머플러", "장갑", "비니"]
    elif band == "1~8도":
        outer = ["코트", "경량패딩/패딩베스트"]
        base = ["니트", "맨투맨 + 이너", "데님/슬랙스"]
        shoes = ["부츠", "로퍼 + 양말"]
        acc = ["머플러", "비니"]
    elif band == "9~16도":
        outer = ["트렌치", "자켓/블레이저", "가디건 레이어드"]
        base = ["긴팔 티", "셔츠", "얇은 니트"]
        shoes = ["스니커즈", "로퍼"]
        acc = ["가벼운 스카프", "에코백"]
    elif band == "17~22도":
        outer = ["가벼운 자켓", "셔츠 아우터"]
        base = ["반팔+가디건", "셔츠", "슬랙스/데님"]
        shoes = ["스니커즈", "로퍼"]
        acc = ["캡/모자", "선글라스(맑음)"]
    elif band == "23~27도":
        outer = ["(옵션) 얇은 셔츠"]
        base = ["반팔", "린넨 셔츠", "얇은 슬랙스/쇼츠"]
        shoes = ["스니커즈", "샌들"]
        acc = ["선글라스", "미니백"]
    else:
        outer = ["(필요 시) 린넨 셔츠"]
        base = ["민소매/반팔", "쇼츠/얇은 팬츠", "통풍 좋은 소재"]
        shoes = ["샌들", "가벼운 스니커즈"]
        acc = ["모자", "양산/선크림"]

    if is_rainy:
        outer = ["바람막이/레인자켓"] + outer
        shoes = ["방수 스니커즈/레인부츠", "미끄럼 방지"]
        acc += ["우산", "방수 가방/커버"]

    if style == "미니멀":
        palette = ["무채색", "뉴트럴(베이지/오프화이트)"]
        fit = ["정돈된 실루엣", "디테일 최소"]
    elif style == "스트릿":
        palette = ["블랙", "포인트 컬러 1개"]
        fit = ["오버핏 상의", "와이드/카고 팬츠"]
        acc += ["캡", "크로스백"]
    elif style == "러블리":
        palette = ["파스텔", "아이보리"]
        fit = ["가디건/니트 레이어드", "스커트/원피스"]
        base += ["(옵션) 롱스커트/원피스"]
    else:
        palette = ["네이비", "브라운", "그레이"]
        fit = ["블레이저/셔츠", "슬랙스"]

    return {"band": band, "outer": outer, "base": base, "shoes": shoes, "acc": acc,
            "palette": palette, "fit": fit}

def local_chatbot_answer(user_msg: str, weather: Weather | None):
    msg = user_msg.strip().lower()

    if any(k in msg for k in ["안녕", "hi", "hello", "반가", "ㅎㅇ"]):
        return "안녕! ‘오늘 뭐 입지?’, ‘비 오는 날 코디’, ‘미니멀 룩 추천’처럼 물어봐줘 :)"

    if any(k in msg for k in ["오늘", "뭐 입", "코디", "추천"]):
        if not weather:
            return "좋아! ‘추천날씨 코디’ 메뉴에서 도시 입력 → 날씨 가져오기 먼저 해줘. 그러면 오늘 온도/비까지 반영해서 추천해줄게."
        return f"{weather.city} {weather.temp_c:.1f}°C 기준으로 추천을 만들 수 있어! ‘추천날씨 코디’에서 스타일을 고르면 더 정확해져."

    if any(k in msg for k in ["색", "컬러", "톤"]):
        return "안전한 시작은 ‘무채색+포인트 1개’야. 선호 색(블랙/베이지/네이비) 알려주면 조합을 더 구체화해줄게."

    if any(k in msg for k in ["출근", "학교", "데이트", "면접"]):
        return "상황이 중요해! 출근/데이트/면접 중 뭐고, 장소(실내/야외)랑 선호 스타일(미니멀/클래식/스트릿/러블리)도 같이 말해줘."

    return "오케이. 키/선호스타일/상황(출근·데이트·학교) 말해주면 더 딱 맞게 추천해줄게!"
```

```python
import streamlit as st
from utils import fetch_weather, outfit_reco

st.title("추천날씨 코디")

if "last_weather" not in st.session_state:
    st.session_state.last_weather = None

city = st.text_input("도시(영문 추천: Seoul, Busan, Daegu)", value="Seoul")
style = st.selectbox("스타일", ["미니멀", "클래식", "스트릿", "러블리"])

col1, col2 = st.columns(2)
with col1:
    if st.button("날씨 가져오기", use_container_width=True):
        try:
            w = fetch_weather(city)
            if not w:
                st.error("도시를 찾지 못했어요. 영문 도시명으로 다시 시도해봐요.")
            else:
                st.session_state.last_weather = w
                st.success(f"{w.city} 날씨 로드 완료!")
        except Exception as e:
            st.error(f"날씨 호출 실패: {e}")

with col2:
    w = st.session_state.last_weather
    if w:
        st.info(f"맑음 {w.city} | 온도{w.temp_c:.1f}°C | 비 {'있음' if w.is_rainy else '낮음'} | 바람 {w.wind_kmh:.1f} km/h")
    else:
        st.warning("아직 날씨가 없어요. ‘날씨 가져오기’를 눌러줘.")

w = st.session_state.last_weather
if w:
    reco = outfit_reco(w.temp_c, w.is_rainy, style)

    a, b, c = st.columns(3)
    with a:
        st.subheader("아우터")
        st.write("• " + "\n• ".join(reco["outer"]))
    with b:
        st.subheader("상·하의")
        st.write("• " + "\n• ".join(reco["base"]))
    with c:
        st.subheader("신발·소품")
        st.write("• " + "\n• ".join(reco["shoes"] + reco["acc"]))

    st.subheader("컬러/실루엣")
    st.write("추천 팔레트:", ", ".join(reco["palette"]))
    st.write("핏 포인트:", ", ".join(reco["fit"]))
```
