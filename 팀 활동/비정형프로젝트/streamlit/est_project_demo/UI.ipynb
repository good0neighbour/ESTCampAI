```python
%%writefile lookXpertM.py
import streamlit as st
import time

# â­ ë°˜ë“œì‹œ ì²« ë²ˆì§¸ Streamlit ëª…ë ¹
st.set_page_config(page_title="LookXpertM", layout="wide")

# -----------------------------
# í—¤ë”
# -----------------------------
st.title("ğŸ‘— LookXpertM")
st.caption("íŒ¨ì…˜ ì¶”ì²œ ì±—ë´‡ â€“ Prototype")

# -----------------------------
# Sidebar â€“ ì‚¬ìš©ì ì •ë³´
# -----------------------------
with st.sidebar:
    st.title("ğŸ‘¤ ì‚¬ìš©ì í”„ë¡œí•„")
    user_name = st.text_input("ë‹‰ë„¤ì„", "íŒ¨ì…”ë‹ˆìŠ¤íƒ€")
    prefer = st.radio("ì„ í˜¸ ìŠ¤íƒ€ì¼", ["ë¯¸ë‹ˆë©€", "ìŠ¤íŠ¸ë¦¿", "ìºì£¼ì–¼", "ë¹„ì¦ˆë‹ˆìŠ¤"])
    st.info(f"{user_name}ë‹˜ ì·¨í–¥ ë°˜ì˜ ì¤‘ì…ë‹ˆë‹¤ ğŸ˜Š")

# ì„¸ì…˜ ì´ˆê¸°í™”
if "history" not in st.session_state:
    st.session_state.history = []

if "favorites" not in st.session_state:
    st.session_state.favorites = []

# -----------------------------
# íƒ­ êµ¬ì„±
# -----------------------------
tab1, tab2, tab3 = st.tabs(["ğŸ‘• ì¶”ì²œ", "ğŸ•‘ ì¶”ì²œ ê¸°ë¡", "â­ ì°œ ëª©ë¡"])

# -----------------------------
# TAB 1 â€“ ì¶”ì²œ
# -----------------------------
with tab1:

    col1, col2, col3 = st.columns(3)

    with col1:
        weather = st.selectbox("â˜€ï¸ ë‚ ì”¨", ["ì„ íƒ", "ì¶”ì›€", "ë³´í†µ", "ë”ì›€", "íë¦¼", "ì¥ë§ˆ"])

    with col2:
        situation = st.selectbox("ğŸ’¼ ìƒí™©", ["ì„ íƒ", "ì¶œê·¼", "ë°ì´íŠ¸", "ìºì£¼ì–¼", "ì—¬í–‰", "ë©´ì ‘"])

    with col3:
        tone = st.selectbox("ğŸ¨ í†¤", ["ì„ íƒ", "ì›œí†¤", "ì¿¨í†¤", "ë‰´íŠ¸ëŸ´"])

    budget = st.slider("ğŸ’° ì˜ˆì‚°", 20000, 300000, 80000)

    if st.button("ì¶”ì²œ ë°›ê¸°", use_container_width=True):
        if "ì„ íƒ" in [weather, situation, tone]:
            st.warning("ëª¨ë“  ê°’ì„ ì„ íƒí•´ì£¼ì„¸ìš”!")
        else:
            with st.spinner("LookXpertM AIê°€ ì½”ë””ë¥¼ ë¶„ì„ ì¤‘ì…ë‹ˆë‹¤..."):
                time.sleep(1.2)

            st.success("ë¶„ì„ ì™„ë£Œ!")

            rec = f"{weather}/{situation}/{tone}/{budget}"
            st.session_state.history.append(rec)

            img_col, text_col = st.columns([1, 2])

            with img_col:
                st.image("https://via.placeholder.com/300x400.png?text=Recommended+Look")

            with text_col:
                st.markdown("### ğŸ·ï¸ ì˜¥ìŠ¤í¬ë“œ ì…”ì¸  + ìŠ¬ë¦¼ ìŠ¬ë™ìŠ¤")
                st.write(f"- ìƒí™©: **{situation}**")
                st.write(f"- í†¤: **{tone}**")
                st.write(f"- ì˜ˆì‚°: **{budget:,}ì› ì´í•˜**")

                if st.button("â­ ì°œí•˜ê¸°"):
                    st.session_state.favorites.append("ì˜¥ìŠ¤í¬ë“œ ì…”ì¸  + ìŠ¬ë¦¼ ìŠ¬ë™ìŠ¤")
                    st.success("ì°œ ëª©ë¡ì— ì¶”ê°€ë˜ì—ˆìŠµë‹ˆë‹¤!")

# -----------------------------
# TAB 2 â€“ ì¶”ì²œ ê¸°ë¡
# -----------------------------
with tab2:
    st.subheader("ğŸ“œ ì´ì „ ì¶”ì²œ ê¸°ë¡")
    st.write(st.session_state.history)

# -----------------------------
# TAB 3 â€“ ì°œ ëª©ë¡
# -----------------------------
with tab3:
    st.subheader("â­ ì°œí•œ ì•„ì´í…œ")
    st.write(st.session_state.favorites)
```
