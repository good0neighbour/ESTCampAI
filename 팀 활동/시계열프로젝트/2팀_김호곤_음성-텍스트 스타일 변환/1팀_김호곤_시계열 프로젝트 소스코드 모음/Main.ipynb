# 런타임 시작 시 먼저 실행

```python
from google.colab import drive
from google.colab import output
import sys

# 경로 설정
work_path = "/content/drive/MyDrive/ESTCampAI/시계열프로젝트"

# 구글 드라이브 연동
drive.mount("/content/drive")
%cd $work_path

# 라이브러리 경로 추가
sys.path.append(work_path)

# 라이브러리 임포트
from AudioTextTransfer import WhisperModel
from TextStyleTransfer import ModelPipeline
```

# 시작 단계

```python
# 모델 생성
audioTransfer = WhisperModel()
styleTransfer = ModelPipeline(model_path = "./data/TextStyleTransfer")
```

# 실행 단계

```python
while True:
  # 오디오-텍스트 변환
  # ./data/GSF03031LJH0_한국어/SW201105KTRNGSF03031LJH0001.pcm
  originalText = audioTransfer.TransferAudio(
      input("오디오 경로 >> ")
  )

  # 지원하지 않는 파일 형식
  if originalText == None:
    continue

  # 문체 스타일 변환
  transferedText = styleTransfer.GenerateText(
      originalText,
      input("스타일 입력 >> "),
      1,
      1024
  )[0]

  # 결과 출력
  print("\n원본 문장: ", originalText)
  print("변환 문장: ", transferedText)

  # 종료 여부
  if input("\n종료하려면 end 입력. 계속하려면 아무 키 입력 >> ") == "end":
    break
  else:
    output.clear()
```

# 종료 단계

```python
# 메모리 해제
del styleTransfer
del audioTransfer

# 종료
print("프로그램 종료")
```
