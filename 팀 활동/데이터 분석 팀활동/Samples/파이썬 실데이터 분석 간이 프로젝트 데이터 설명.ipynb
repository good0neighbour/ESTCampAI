> **시나리오**


```A고등학교에서 근무하고 있는 학생 주임 선생님은 최근 들어 골머리를 앓고 있다. 학생들의 일탈 행위가 지속적으로 지속적으로 증가하고 있는 상황이다. 단순한 체벌 및 훈육으로 제대로 학생 지도를 할 수 없다고 판단하여, 학생들의 스트레스를 감소시켜 근본적으로 학생들의 일탈 행위를 막으려고 한다.```<br/>
```이에 각 학생들의 스트레스 요인이 기입된 데이터를 활용하여, 알맞은 정책 도입함으로써 근본적으로 학생들의 일탈 행위를 방지하고자 한다.```


> **데이터 살펴보기**

* 설문을 통해 학생들이 스트레스 정도와 이와 관련된 항목들을 수집했다.
* 데이터 명세 ↓

> **데이터 명세**

* 불안수준 : 불안함을 느끼는 정도  (0 ~ 21) (GAD-7)
* 자존감 : 스스로를 사랑하는 정도 (0 ~ 30) (로젠버그 자존감 척도)
* 정신과병력 : 정신과 관련 병을 진단받았는지 (Fasle : 진단 X,  True : 진단 O) 
* 우울수준 : 우울감을 느끼는 정도 (0 ~ 27) (PHQ-9)
* 두통 : 두통을 얼마나 자주 겪는지 (0 : 없음, 1 : 거의 없음, 2 : 가끔 겪음, 3 : 종종 겪음, 4 : 자주 겪음 ,5 : 매우 자주 겪음)
* 수면질 : 수면의 질이 좋은지 (0 : 극도로 나쁨, 1 : 매우 나쁨, 2 : 약간 나쁨, 3 : 보통, 4 : 좋음, 5 : 매우 좋음)
* 삶의질 : 전반적인 삶의 만족도 (0 : 극도로 나쁨, 1 : 매우 나쁨, 2 : 약간 나쁨, 3 : 보통, 4 : 좋음, 5 : 매우 좋음)
* 학업성취도 : 전반적인 학업 성취도 (0 : 극도로 나쁨, 1 : 매우 나쁨, 2 : 약간 나쁨, 3 : 보통, 4 : 좋음, 5 : 매우 좋음)
* 학업부담도 : 학업에 부담을 느끼는 정도 (0 : 없음, 1 : 거의 없음, 2 : 가끔 겪음, 3 : 종종 겪음, 4 : 자주 겪음 ,5 : 매우 자주 겪음)
* 교사-학생관계도 : 교사와의 관계도 (0 : 극도로 나쁨, 1 : 매우 나쁨, 2 : 약간 나쁨, 3 : 보통, 4 : 좋음, 5 : 매우 좋음)
* 진로고민도 : 미래 진로에 대해 고민하는 정도 (0 : 없음, 1 : 거의 없음, 2 : 가끔 겪음, 3 : 종종 겪음, 4 : 자주 겪음 ,5 : 매우 자주 겪음)
* 사회적지원 : 사회로부터 충분히 지원을 받고 있다고 생각하는지 (0 : 없음, 1 : 거의 없음, 2 : 가끔 겪음, 3 : 종종 겪음, 4 : 자주 겪음 ,5 : 매우 자주 겪음)
* 압박감정도 : 동급생들에게 압박감을 얼마나 느끼는지 (0 : 없음, 1 : 거의 없음, 2 : 가끔 겪음, 3 : 종종 겪음, 4 : 자주 겪음 ,5 : 매우 자주 겪음)
* 과외수준 : 과외를 진행하는 빈도
* 괴롭힘 : 동급생으로부터 괴롭힘을 당하는 정도 (0 : 없음, 1 : 거의 없음, 2 : 가끔 겪음, 3 : 종종 겪음, 4 : 자주 겪음 ,5 : 매우 자주 겪음)
* 스트레스 : 스트레스를 느끼는 정도 (0 : 낮음, 1 : 중간, 2 : 높음)
