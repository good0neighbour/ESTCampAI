### matplotlib
- 데이터 시각화를 도와주는 라이브러리
- 데이터 시각화 : 분산된 데이터를 도표나 차트로 정리하는 행위
- 데이터 시각화를 통해 데이터의 특징을 한눈에 파악 가능
- 분석 결과를 상대방에게 효과적으로 전달 가능

```python
# matplotlib 불러오기
import matplotlib.pyplot as plt
# matplotlib 아래에 있는 pyplot 호출

from matplotlib import pyplot as plt
```

```python
# 라이브러리 -> import하기 전 라이브러리를 설치하는 과정이 필요
# anaconda를 통해 jupyter notebook 사용
# 아나콘다가 데이터 분석에 자주 사용되는 라이브러리(numpy, pandas, matplotlib)를 다 설치해줬음
```

```python
# 그래프 그래는 순서
# 1. 데이터 준비
# 2. 그래프를 그릴 도화지 준비
# 3. 그래서 설정
# 4. 그래프 출력
```

```python
# 선그래프 그려보기.
# 1. 데이터 준비
# x, y(필수) 데이터 준비 → x를 안 넘겨주는 경우, 0부터 시작하는 x값을 자동으로 생성
y = [2, 4, 6]

# 2. 도화지 준비 생략.

# 3. 그래프 설정
plt.plot(y) # 선그래프를 그리는 함수

# 4. 그래프 출력
plt.show()
```

```python
# x, y 데이터 모두 줘보기
# 1. 데이터 준비
# y는 위에 정의한 데이터 그대로
x = [1, 2, 3]
# 2. 도화지 준비 생략

# 3. 그래프 설정
plt.plot(x, y)
# 4. 그래프 출력
plt.show()
```

```python
# matplotlib이 가진 굉장한 장점 : 커스터마이징이 잘 됨
# 먼저 그래프를 그리는 도화지를 커스텀해보자.

# figure를 통해서 설정 가능
# 1) figsize : 도화지의 크기 (가로, 세로)
# 2) facecolor : 도화지 배경색
plt.figure(figsize = (3, 3), facecolor = 'green')

plt.plot(x, y)

plt.show()
```

```python
# 그래프 내부 요소를 커스텀
# line style(ls) : 선 꾸며주기
# line width(lw) : 선 굵기
# color : 선의 색깔
# marker : marker 타입 설정
# markerfacecolor(mfc) : marker 내부 색상
# markeredgecolor(mec) : marker 테두리 색상

# 그래프 설정할 때 커스텀 해주면 됨
plt.plot(x, y, ls = ':', marker = 'o', lw = 4, color = 'red', mfc = 'blue', mec = 'black', mew = 2)

plt.show()
```

```python
# numpy import
import numpy as np
x = np.arange(1, 9)
y = [5, 2, 3, 6, 8, 7, 5, 7]

# 직접 그래프를 그려보기.
plt.plot(x, y, ls = '--', marker = 'o', lw = 2, color = 'green')
plt.show()
plt.plot(x, y, ls = ':', marker = 'D', lw = 2, color = 'red', mfc = 'blue', mec = 'red')
plt.show()
```

```python
# 여러개의 그래프를 한번에 출력하는 방법, 같은 도화지에 출력함
# subplot : 표를 만들어서 원하는 위치에 그래프를 여러개 출력 가능

# subplot(세로, 가로, 다음에 올 그래프의 위치)
plt.figure(figsize = (12, 4.5))
plt.subplot(1, 2, 1)
plt.plot(x, y, ls = '--', marker = 'o', lw = 2, color = 'green')
plt.subplot(1, 2, 2)
plt.plot(x, y, ls = ':', marker = 'D', lw = 2, color = 'red', mfc = 'blue', mec = 'red')
plt.show()
```

```python
# 격자, x, y축 범위 설정

# 그래프 생성
plt.plot(x, y, color = 'plum', ls = '-.', marker = 's', mfc = 'white')

# 1. x, y 축 범위 -> limitation -> lim
plt.xlim(-4, 10)
plt.ylim(-2, 12)

# 2. 좌표평면에 격자 무늬 그리기
plt.grid()

plt.show()
```

```python
x = [1, 2, 3]

y1 = [1, 2, 3]
y2 = [2, 4, 6]
y3 = [3, 6, 9]
# 선 그래프를 한 좌표평면 상에 여러개 그리기.

plt.figure(figsize = (4, 3))

# 그래프 생성
plt.plot(x, y1, label = 'y1')
plt.plot(x, y2, label = 'y2')
plt.plot(x, y3, label = 'y3')

# legend()
# plot에서 이름표(label)를 먼저 설정해주기.
plt.legend()

# pyplot 설정 값이 모여있는 딕셔너리
plt.rcParams['font.family'] = 'Gulim'

# 그래프 제목 붙여주기.
plt.title('첫 번째 그래프')
# matplotlib은 기본적으로 한글 출력 안 됨. → 한글 출력할 수 있는 폰트 설정이 안 돼서

# 그래프 파일 저장
plt.savefig('./data/chart.png', dpi = 300) # 저장 경로+파일명, 화질

plt.show()
```
