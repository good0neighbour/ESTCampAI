```python
# )    "00 00 00"  .

# 
time = int(input("  >>"))

# 
hour = time // 3600
hourInSecond = hour * 3600

# 
minute = (time - hourInSecond) // 60
minuteInSecond = minute * 60

# 
second = time % 60

#  
print("{} {} {}".format(hour, minute, second))

# 
print(f"\n\n{hour}() * 3600 = {hourInSecond}() | {minute}() * 60 = {minuteInSecond}() | {second}()")
print(f"{hourInSecond} + {minuteInSecond} + {second} = {hourInSecond + minuteInSecond + second}")
```
