## 반복문 실습 과제

### 1. While문 활용


```python
while True:
    userID = input("아이디를 입력하세요 >>")
    userPW = input("비밀번호를 입력하세요 >>")
    if userID == "Hello" and userPW == "1234":
        print("로그인성공")
        break
    else:
        print("아이디와 비밀번호가 잘못되었습니다.")
```

### 2. for문 활용


```python
start = int(input("첫 번째 정수 입력 >>"))
end = int(input("두 번째 정수 입력 >>"))
for i in range(start, end + 1):
    print(i, end = " ")
```

### 3. for문 활용


```python
num = int(input("정수 입력 >>"))
print(f"{num}의 약수 :", end = " ")
for i in range(1, num + 1):
    if num % i == 0:
        print(i, end = " ")
```

### 4. for문 활용


```python
list2 = [4, 5, 2, 1, 99, 15, 2, 7, 27]
result = 0
for i in list2:
    if result < i:
        result = i
print(result)
```
