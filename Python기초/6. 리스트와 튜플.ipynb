### 리스트
- 파이썬의 자료형(정수, 실수, 문자열 등) 중 하나
- 순서가 있는 수정 가능한 객체의 집합
- 대괄호로 작성되어지며 내부의 값을 콤마로 구분한다
- 추가, 수정, 삭제 가능

```python
# 리스트 만들어보기
a = []
b = [1, 2, 3]
c = ['My', 'name']
d = [1, 2, 'My', 'name'] # 문자와 숫자를 섞어서 생성도 가능
e = [1, 2, ['My', 'name']] # 이중리스트 생성도 가능
print(a)
print(type(a))
print(b)
print(type(b))
print(c)
print(type(c))
print(d)
print(type(d))
print(e)
print(type(e))
```

### 리스트의 인덱싱
- 리스트명[인덱스번호]

```python
list1 = [2, 5, 7, 9, 10]
```

```python
print(list1[3]) # 정방향
print(list1[-2]) # 역방향
print(list1[1] + list1[-1])
```

```python
list2 = [1, 2, 3, ['a', 'b', 'c']]
temp = list2[3]
print(temp)
print(list2[3][1])
```

#### 리스트의 슬라이싱
- 리스트명[시작인덱스번호:끝인덱스번호+1]

```python
list3 = [0, 2, 4, 6, 8, 10]
```

```python
# 2, 4, 6 슬라이싱해오기
print(list3[1:4])
# 전체를 슬라이싱해오기
print(list3[:])
```

```python
# 리스트의 연산
list4 = [1, 2, 3]
list5 = [3, 4, 5, 6]
list4 + list5
```

```python
# 리스트의 값 추가 -> append(값) : 맨 뒤에 값을 추가할 수 있다
list6 = [0, 1, 2, 3, 4, 5]
list6.append(6)
print(list6)
```

```python
f_list = ['사과', '포도', ['수박', '멜론'], '복숭아', '딸기', '오렌지']
choice_list = []
choice_list.append(f_list[0])
choice_list.append(f_list[-1])
choice_list.append(f_list[2][0])
print(choice_list)
```

```python
# 값 추가하는 함수 -> insert(인덱스위치, 값) : 인덱스 위치에 값을 추가
list7 = [0, 2, 4, 6]
list7.insert(1, 3)
print(list7)
```

```python
# 정수 형태의 인덱스 값과 노래 제목, 가수명도 각각 입력받은 후 변수에 저장하고, insert 함수를 이용하여 music_list에 새로운 값을 추가하시오.
loc = int(input("인덱스위치입력 >>"))
song = input("노래제목입력 >>")
singer = input("가수입력 >>")

music_list = [['Butter','방탄'],['헤픈우연','헤이즈'],['라일락','아이유']]
music_list.insert(loc, [song, singer])
print(music_list)
```

```python
# 리스트 값 수정 -> 리스트명[수정이필요한 인덱스번호] = 값
list8 = [0, 1, 3, 5, 7]
list8[-2] = 10
print(list8)
```

```python
list8[2:4] = [7]
print(list8)
```

```python
array = [1, 2, 3, 4, 5]

array[-1] = 7
print(array)

array[1] = ['a', 'b', 'c']
print(array)

array[3:5] = ['d', 'e', 'f', 'g']
print(array)

array[0:1] = ['h', 'i', 'j']
print(array)
```

```python
# 리스트 삭제 -> del 리스트명[인덱스번호], del 리스트명[시작인덱스:끝인덱스+1]
list8 = [0, 1, 7, 7]
del list8[2]
print(list8)
del list8[1:]
print(list8)
```

```python
# 리스트 삭제 -> remove('값')
list9 = ['a', 'b', 'c', 'd', 'e', 'c']
list9.remove('c')
print(list9)
```

```python
# 리스트의 오름차순 정렬 -> sort()
list10 = [9, 77, 13, 51, 100, 3]
list10.sort()
print(list10)
```

```python
# 리스트의 역순 정렬 -> reverse()
list10 = [9, 77, 13, 51, 100, 3]
list10.reverse()
list10
```

```python
# 내림차순으로 정렬해보기
list10 = [9, 77, 13, 51, 100, 3]
list10.sort()
list10.reverse()
print(list10)
```

```python
list10 = [9, 77, 13, 51, 100, 3]
list10.sort(reverse = True)
print(list10)
```

```python
# 리스트의 찾고자하는 값의 위치 확인하기 -> index()
list11 = ['a', 'b', 'c', 'd', 'e']
list11.index('d')
```

```python
# 리스트의 찾고자하는 값 개수 확인하기 -> count()
list12 = ['a', 'b', 'c', 'c', 'c']
list12.count('c')
```

```python
# 리스트의 데이터의 개수를 확인하기 -> len()
len(list12)
```

### 튜플
- 파이썬의 자료형 중 하나
- 순서가 있는 집합
- 소괄호로 작성되어지며 내부의 값을 콤마로 수분한다.
- 추가, 수정, 삭제 불가능

```python
a = ()
b = (1, 2, 3)
c = ("My", "name")
d = (1, 2, "My", "name")
e = (1, 2, ("My", "name"))
print(a)
print(type(a))
print(b)
print(type(b))
print(c)
print(type(c))
print(d)
print(type(d))
print(e)
print(type(e))
```

```python
tuple1 = (0, 1, 2, 3, ('a', 'b', 'c'), 4)
# 인덱싱
print(tuple1[4])
# 슬라이싱
print(tuple1[3:])
```

### 멤버연산자


```python
# 멤버연산자, 조건문, 인덱스번호 찾는 함수 활용하기
lst = ["딸기", "바나나", "수박", "체리", "포도"]
text = input("검색할 문자를 입력하세요 >>")
if text in lst:
    print(f"{text}는 리스트에 {lst.index(text)}번째 인덱스에 들어있습니다.")
else:
    print(f"{text}는 리스트에 들어있지 않습니다.")
```

```python
# 사용자가 입력한 글자가 몇 개 포함되어 있는지 출력
s = "Hi, My name is Mihee"

char = input("검색할 문자를 입력하세요 :")
num = s.count(char)

if 0 < num:
    print(f"'{char}'는 {num}번 들어있습니다.")
else:
    print(f"'{char}'는 문자열에 들어있지 않습니다.")
```
