### (Dictionary)
-     
-    Key    Value     .
-  .
-  ({,})   (,) .
- Key, Value     ():) .


```python
a = {}
b = {"name": "DMX"}
c = {1: 5, 2: 6}
d = {'': ["DMX"]} #           .
print(a)
print(type(a))
print(b)
print(type(b))
print(c)
print(type(c))
print(d)
print(type(d))
```

```python
info = {'name': 'DMX', 'age': 20, 'number': '1-800-225-5324'}
print(info)
```

```python
#   
info['birth'] = '12/25'
print(info)
```

```python
dic_test = {"": ""}
dic_test[""] = ""
dic_test[""] = "2025.08.01"
print(dic_test)
```

```python
#  value  -> [key]
print(info["name"])
print(info["birth"])
```

```python
#  value  -> .get(key)
print(info.get('name'))
print(info.get('birth'))
```

```python
print(info.get('')) #  ->    None , X
print(info[''])
```

```python
print(info.keys())
print(info.values())
```

```python
for key in info.keys():
    print(key)
```

```python
for value in info.values():
    print(value)
```

```python
for k, v in info.items():
    print(k, v)
```

```python
#    -> del [key]
del info['name']
print(info)
```

```python
del info['age']
print(info)
```

```python
#      -> .clear()
info.clear()
print(info)
```
