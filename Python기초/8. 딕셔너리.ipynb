### 딕셔너리(Dictionary)
- 단어 그대로 해석하면 사전이라는 의미
- 변할 수 없는 Key값과 변할 수 있는 Value를 한 쌍으로 가지고 있는 자료형이다.
- 순서가 없다.
- 기호는 중괄호({,})로 표현하며 구분은 콤마(,)를 사용한다.
- Key, Value 값을 한쌍으로 표현할 때 콜론():)을 사용한다.


```python
a = {}
b = {"name": "DMX"}
c = {1: 5, 2: 6}
d = {'이름': ["DMX"]} # 딕셔너리 키값에 리스트 형식의 변할 수 있는 자료형을 지정할 수 없다.
print(a)
print(type(a))
print(b)
print(type(b))
print(c)
print(type(c))
print(d)
print(type(d))
```

```python
info = {'name': 'DMX', 'age': 20, 'number': '1-800-225-5324'}
print(info)
```

```python
# 딕셔너리에 값 추가하기
info['birth'] = '12/25'
print(info)
```

```python
dic_test = {"노래제목": "네모의꿈"}
dic_test["가수"] = "아이유"
dic_test["날짜"] = "2025.08.01"
print(dic_test)
```

```python
# 딕셔너리의 value값 가져오기 -> 딕셔너리명[key]
print(info["name"])
print(info["birth"])
```

```python
# 딕셔너리의 value값 가져오기 -> 딕셔너리명.get(key)
print(info.get('name'))
print(info.get('birth'))
```

```python
print(info.get('성별')) # 함수 -> 없는 키값을 호출하면 None 반환, 오류X
print(info['성별'])
```

```python
print(info.keys())
print(info.values())
```

```python
for key in info.keys():
    print(key)
```

```python
for value in info.values():
    print(value)
```

```python
for k, v in info.items():
    print(k, v)
```

```python
# 딕셔너리 값 삭제 -> del 딕셔너리명[key]
del info['name']
print(info)
```

```python
del info['age']
print(info)
```

```python
# 딕셔너리 내 모든 정보 삭제 -> 딕셔너리명.clear()
info.clear()
print(info)
```
