커서를 셀 안에 두고 편집할 수 있는 상태 : Edit Mode<br>
커서 밖에 두고 단축키를 사용할 수 있는 상태 : Command Mode

### 단축키
 - 실행 단축키 : Ctrl+Enter(해당하는 셀 실행), Alt+Enter(실행 후 다음 셀 커서 위치), Shit+Enter(실행 후 다음 셀 생성)
 - 셀 삭제 : dd
 - 아래쪽 셀 추가 : b
 - 위쪽 셀 추가 : a
 - 셀을 Markdown으로 변경 : m
 - 셀을 code로 변경 : y
