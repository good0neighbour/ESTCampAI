### 조건문
- 상황에 따라 실행의 흐름을 다르게 하는 문법


```python
if 조건식:
    실행문장
```

```python
if True:
    print("실행문장 실행")
print("if문 밖에 있는 실행문장")
```

```python
if False:
    print("실행문장 실행")
print("if문 밖에 있는 실행문장")
```

```python
# 조건문을 사용하여 변수 money가 10000 이상이면 "택시를 탄다"를 출력하시오.
money = int(input("얼마 있니?"))
if 10000 <= money:
    print("택시를 탄다.")
```

```python
# 변수 score에 점수를 입력 받아서 학점 부여.
score = int(input("점수 입력 >>"))
if 90 <= score:
    rank = "A"
elif 80 <= score:
    rank = "B"
elif 70 <= score:
    rank = "C"
elif 60 <= score:
    rank = "D"
else:
    rank = "F"
print(f"{score}점은 {rank}학점 입니다.")
```

```python
# 변수 num1과 num2에 숫자를 입력 받아 크기를 비교하시오.
num1 = int(input("첫 번째 정수 입력 >>"))
num2 = int(input("두 번째 정수 입력 >>"))
if num1 > num2:
    print("첫 번째 정수가 더 큽니다.")
elif num1 < num2:
    print("두 번째 정수가 더 큽니다.")
else:
    print("두 수가 똑같습니다.")
```

```python
height = float(input("키"))
age = int(input("나이"))
price = 0
if 120.0 < height:
    print("탈 수 있음.")
    if 12 > age:
        price += 500
    elif 18 >= age:
        price += 1000
    else:
        price += 3000
    if input("사진(Y/N)") == "Y":
        price += 300
    print(f"총비용 {price}")
else:
    print("탈 수 없음.")
```

```python
# 자판기 프로그램

#자판기 디자인
print('''
================================================
================---- 자판기 ----================
================================================
== 1.콜라 == 2.사이다 ==  3.물  ========insert===
==--------==----------==--------=======------===
==   600  ==    800   ==  1000  ======= O00원===
=======================================------===
================================================
==++++++++++++++++++++++++++++++++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
==+++++++++++++++              +++++++++++++++==
==+++++++++++++++    ESTCAMP   +++++++++++++++==
==+++++++++++++++              +++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
================================================
================================================
  ===                                      ===
  ===                                      ===
''')

# 입력
coin = int(input("insert coin >>"))
menu = int(input("메뉴 선택 >>"))

# 금액 확인
if 1 == menu:
    if 600 <= coin:
        coin -= 600
    else:
        print("잔액부족")
elif 2 == menu:
    if 800 <= coin:
        coin -= 800
    else:
        print("잔액부족")
elif 3 == menu:
    if 1000 <= coin:
        coin -= 1000
    else:
        print("잔액부족")
else:
    print("잘못된 메뉴")

# 잔액 환전
refund_1000 = coin // 1000
coin %= 1000
refund_500 = coin // 500
coin %= 500
refind_100 = coin // 100
coin %= 100

# 잔돈 출력
print(f"잔돈 >> 1000원 {refund_1000}개, 500원 {refund_500}개, 100원 {refind_100}개")

# 자판기가 먹은 돈
if coin > 0:
    print(f"자판기가 먹은 돈 >> {coin}")
```
