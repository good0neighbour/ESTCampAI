### 반복문
- 프로그램 내에서 똑같은 명령을 일정 횟수만큼 반복하여 수행하도록 제어하는 명령문
- 반복문 종류 : while, for
- while : 반복횟수가 명확하지 않을 때
- for : 반복횟수가 명확할 때


```python
# 1부터 3까지 while문을 통해서 출력하기.
num = 1
while num <= 3:
    print(num)
    num = num + 1
```

```python
# 0부터 10까지 while문을 통해서 출력하기.
num = 0
while num <= 10:
    print(num)
    num += 1
```

```python
# while문 사용해서 파이썬최고라는 문자열을 13번 출력해라.
i = 0
while i < 13:
    print("파이썬최고")
    i += 1
```

```python
# 무한반복 -> 조건식이 무조건 True, 멈추기위해서는 break 기능을 사용한다
while True:
    print("무한루프")
    break # 반복문을 나가는 기능
```

```python
# 무한 반복을 활용해서 1~3까지 출력하기
num = 1
while True:
    print(num)
    num += 1
    if num > 3:
        break
```

```python
# 두 개의 정수를 입력 받아서 더하는 코드를 작성하시오.
# 단, 두 개의 정수가 0이 들어올 때까지 반복한다.
a = 0
b = 0
while True:
    a = int(input("첫 번째 정수 입력 >>"))
    b = int(input("두 번째 정수 입력 >>"))
    if a == 0 and b == 0:
        print("프로그램이 종료되었습니다.")
        break
    else:
        print(f"두 정수의 합 : {a + b}")
```

```python
# 1. 현재 몸무게와 목표 몸무게를 입력받고 주차 별 감량 몸무게를 입력 받으세요.
# 2. 목표 몸무게를 달성하면 축하한다는 문구를 출력하고 입력을 멈추세요.
currentWeight = float(input("현재 몸무게 :"))
targetWeight = float(input("목표 몸무게 :"))
weightLoss = 0.0
week = 1
while True:
    weightLoss = float(input(f"{week}주차 감량 몸무게 :"))
    currentWeight -= weightLoss
    if currentWeight <= targetWeight:
        print(f"{currentWeight}kg 달성. 축하합니다.")
        break
    else:
        week += 1
```

### for문


```python
list_food = ["햄버거", "치킨", "피자"]
for food in list_food:
    print(food)
```

```python
tuple_food = ("햄버거", "치킨", "피자")
for f in tuple_food:
    print(f)
```

```python
s = "안녕하세요"
for st in s:
    print(st)
```

```python
# 5명에 대한 정보처리기사 자격증 시험 점수가 리스트에 담겨있습니다. 이때 각 점수가 함격 점수인지 불합격 점수인지 판별하여 출력하시오.
score_list = [90, 45, 70, 60, 55]
i = 1
for score in score_list:
    if 60 <= score:
        print(f"{i}번 학생은 합격입니다.")
    else:
        print(f"{i}번 학생은 불합격입니다.")
    i += 1
```

```python
for i in range(1, 11, 3):
    print(i, end = " ")
```

```python
for i in range(11):
    print(i)
```

```python
# for문을 이용하여 97부터 77출력하시오.
for i in range(97, 76, -1):
    print(i, end = " ")

print()

# for문을 이용하여 23부터 39출력하시오.
for i in range(23, 40):
    print(i, end = " ")
```

```python
# 0부터 100사이의 숫자 중 3의 배수인 값들의 합을 출력.
sum = 0
for i in range(0, 100):
    if i % 3 == 0:
        sum += i
print(sum, "\n")

# for문을 이용하여 구구단 2단을 출력하시오.
for i in range(1, 10):
    print(f"2 * {i} = {2 * i}")
```
