```python
# 1 3 while  .
num = 1
while num <= 3:
    print(num)
    num = num + 1
```

```python
# 0 10 while  .
num = 0
while num <= 10:
    print(num)
    num += 1
```

```python
# while    13 .
i = 0
while i < 13:
    print("")
    i += 1
```

```python
#  ->   True,  break  
while True:
    print("")
    break #   
```

```python
#    1~3 
num = 1
while True:
    print(num)
    num += 1
    if num > 3:
        break
```

```python
#        .
# ,    0   .
a = 0
b = 0
while True:
    a = int(input("    >>"))
    b = int(input("    >>"))
    if a == 0 and b == 0:
        print(" .")
        break
    else:
        print(f"   : {a + b}")
```

```python
# 1.           .
# 2.        .
currentWeight = float(input("  :"))
targetWeight = float(input("  :"))
weightLoss = 0.0
week = 1
while True:
    weightLoss = float(input(f"{week}   :"))
    currentWeight -= weightLoss
    if currentWeight <= targetWeight:
        print(f"{currentWeight}kg . .")
        break
    else:
        week += 1
```

### for


```python
list_food = ["", "", ""]
for food in list_food:
    print(food)
```

```python
tuple_food = ("", "", "")
for f in tuple_food:
    print(f)
```

```python
s = ""
for st in s:
    print(st)
```

```python
# 5       .         .
score_list = [90, 45, 70, 60, 55]
i = 1
for score in score_list:
    if 60 <= score:
        print(f"{i}  .")
    else:
        print(f"{i}  .")
    i += 1
```

```python
for i in range(1, 11, 3):
    print(i, end = " ")
```

```python
for i in range(11):
    print(i)
```

```python
# for  97 77.
for i in range(97, 76, -1):
    print(i, end = " ")

print()

# for  23 39.
for i in range(23, 40):
    print(i, end = " ")
```

```python
# 0 100   3    .
sum = 0
for i in range(0, 100):
    if i % 3 == 0:
        sum += i
print(sum, "\n")

# for   2 .
for i in range(1, 10):
    print(f"2 * {i} = {2 * i}")
```
